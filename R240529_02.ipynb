```python
# 출근 시간 데이터를 모두 추출

morning_data = data[(pd.to_datetime(data.timestamp).dt.hour < 9) & 
                                (pd.to_datetime(data.timestamp).dt.hour >5)]
morning_data
```

```python
# 출근 시간 데이터를 '역(station_code)' 기준으로 그룹화하여 더하기
morning_station_sum = morning_data.groupby("station_code").sum()
morning_station_sum
```

```python
## 출근 시간 데이터와 역 정보 합치기
morning_joined_data = morning_station_sum.join(station_info)
morning_joined_data
```

```python
## 출근 시간 승차 인원
morning_seoul_in = folium.Map(location=[37.56681, 126.9782], zoom_start = 12) 
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_in']],gradient={0.3: 'blue', 0.6: 'yellow', 1.0: 'red'}).add_to(morning_seoul_in) 
morning_seoul_in
```

```python
morning_seoul_out = folium.Map(location=[37.56681, 126.9782], zoom_start = 12) 
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_out']],gradient={0.3: 'yellow', 0.6: 'green', 1.0: 'blue'}).add_to(morning_seoul_out) 
morning_seoul_out
```

```python
evening_data = data[(pd.to_datetime(data.timestamp).dt.hour < 22) & 
                                (pd.to_datetime(data.timestamp).dt.hour >17)]
evening_data
```

```python
evening_station_sum = evening_data.groupby("station_code").sum()
evening_station_sum
```

```python
evening_joined_data = evening_station_sum.join(station_info)
evening_joined_data
```

```python
evening_seoul_in = folium.Map(location=[37.56681, 126.9782], zoom_start = 12) 
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_in']],gradient={0.3: 'blue', 0.6: 'yellow', 1.0: 'red'}).add_to(evening_seoul_in) 
evening_seoul_in
```

```python
evening_seoul_out = folium.Map(location=[37.56681, 126.9782], zoom_start = 12) 
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_out']],gradient={0.3: 'yellow', 0.6: 'green', 1.0: 'blue'}).add_to(evening_seoul_out) 
evening_seoul_out
```
